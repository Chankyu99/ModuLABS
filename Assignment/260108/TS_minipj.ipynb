```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv('AirPassengers.csv')
data.head()
```

```python
data.drop('Month', axis=1, inplace=True)
data
```

```python
plt.figure()
plt.plot(data)
plt.show()
```

```python
data_log = np.log(data)
plt.figure()
plt.plot(data_log)
plt.show()
```

```python
data_log_diff = data_log.diff()
plt.figure()
plt.plot(data_log_diff)
plt.show()
```

```python
data_final = data_log.diff(12)
data_final.dropna(inplace=True)

plt.figure()
plt.plot(data_final)
plt.show()
```

```python
def adf_test(x): # adfê²€ì •ì„ ìœ„í•œ í•¨ìˆ˜ ì •ì˜
    stat, p_value, lags, nobs, crit, icb = adfuller(x) # adfuller í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ë©´ 6ê°€ì§€ outputì´ ë‚˜ì˜µë‹ˆë‹¤.
    print('ADF statistics')
    print(stat) # ADF ê²€ì • í†µê³„ëŸ‰ìž…ë‹ˆë‹¤.
    print('P-value')
    print(p_value) # P-valueìž…ë‹ˆë‹¤.
```

```python
adf_test(data_final)
```

```python
# â‘¥-1 : ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë° ë¶ˆëŸ¬ì˜¤ê¸°
from tsfresh.examples.robot_execution_failures import download_robot_execution_failures, load_robot_execution_failures

# robot exection failures ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
download_robot_execution_failures()
timeseries, y = load_robot_execution_failures()
```

```python
# â‘¥-2 : ë°ì´í„°ì…‹ í™•ì¸(timeseriesì™€ y í™•ì¸)
timeseries # id, timeì„ ì œì™¸í•˜ê³  6ê°œ columnì˜ Xë³€ìˆ˜ê°€ ë‚˜íƒ€ë‚©ë‹ˆë‹¤.
```

```python
y # true/falseê°€ ì œì‹œë˜ì–´ ìžˆìŠµë‹ˆë‹¤.
```

```python
# â‘¥-3 : ë°ì´í„°ì…‹ ê°€ê³µ
def custom_classification_split(x,y,test_size=0.3): # ìˆ˜ì—… ë•Œ ë³´ì•˜ë˜ custom split í•¨ìˆ˜

    num_true = int(y.sum()*test_size)
    num_false = int((len(y)-y.sum())*test_size)

    id_list = y[y==False].head(num_false).index.to_list() + y[y==True].head(num_true).index.to_list()

    y_train = y.drop(id_list).sort_index()              # yì—ì„œ id_listë¥¼ dropí•©ë‹ˆë‹¤.
    y_test = y.loc[id_list].sort_index()                # 19~38, 1~6ì´ í•©ì³ì§„ ë¦¬ìŠ¤íŠ¸ë¥¼ ì •ë ¬í•©ë‹ˆë‹¤.
    X_train = x[~x['id'].isin(id_list)] # ëŒ€ê´„í˜¸ ì•ˆì— ìžˆëŠ” timeseriesì˜ idì™€ id_listê°€ ì¼ì¹˜í•˜ëŠ” ê²ƒë§Œ ì‚¬ìš©í•˜ê³  ë¬¼ê²°í‘œì‹œëŠ” ì•ˆì— ì¡°ê±´ì´ í¬í•¨ë˜ì–´ ìžˆì§€ ì•ŠëŠ”ê²ƒë§Œ ì‚¬ìš©í•˜ëŠ” ê²ƒìž…ë‹ˆë‹¤.
    X_test = x[x['id'].isin(id_list)]   #

    return X_train, y_train, X_test, y_test
```

```python
# â‘¥-4 : ë°ì´í„°ì…‹ ë‚˜ëˆ„ê¸°
# test setì˜ ë¹„ìœ¨ì„ 0.25ë¡œ ì§€ì •í•´ì£¼ì„¸ìš”.
X_train, y_train, X_test, y_test = custom_classification_split(timeseries, y, test_size=0.25)
```

```python
from tsfresh import extract_features
from tsfresh.feature_extraction import EfficientFCParameters
from tsfresh.utilities.dataframe_functions import impute

settings = EfficientFCParameters()
```

```python
# â‘¥-5 : Feature extraction
comprehensive_features_train = extract_features(
                                    X_train,
                                    column_id="id",
                                    column_sort="time",
                                    default_fc_parameters=settings
                                          )
```

```python
comprehensive_features_test = extract_features(
                                    X_test,
                                    column_id="id",
                                    column_sort="time",
                                    default_fc_parameters=settings
                                          )
```

```python
# â‘¥-6 : impute
impute(comprehensive_features_train)
```

```python
impute(comprehensive_features_test)
```

```python
# â‘¥-7 : RandomForest ì‚¬ìš© ë° score í™•ì¸
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators = 10, max_depth = 3)

rf_clf.fit(comprehensive_features_train, y_train)
```

```python
# testì…‹ì— ì í•©í•´ë³¸ ê²°ê³¼ ì•½ 0.66ì˜ ì •í™•ë„ë¥¼ ë³´ì´ë©° ë¶ˆë§Œì¡±ìŠ¤ëŸ¬ìš´ ê²°ê³¼ë¥¼ í™•ì¸í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.
rf_clf.score(comprehensive_features_test, y_test)
```

```python
# â‘¥-7 : XGBoost ì‚¬ìš© ë° score í™•ì¸
import xgboost as xgb
xgb_clf = xgb.XGBClassifier(n_estimators = 10, max_depth = 3)

# XGBoost ëª¨ë¸ì„ í•™ìŠµì‹œí‚µë‹ˆë‹¤.
xgb_clf.fit(comprehensive_features_train, y_train)
```

```python
# XGBoost ëª¨ë¸ì˜ scoreë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
xgb_clf.score(comprehensive_features_test, y_test)
```

```python
# â‘¥-8 : XGBoost plot_importance í™•ì¸ ë° ì‹œê°í™”
# plot_importanceë¥¼ í†µí•´ ì–´ë–¤ ë³€ìˆ˜(íŠ¹ì§•)ê°€ ê°€ìž¥ ìœ ì˜ë¯¸í•œì§€ í™•ì¸í•©ë‹ˆë‹¤.
xgb.plot_importance(xgb_clf, importance_type = 'gain')
plt.show()
```

```python
sum(xgb_clf.feature_importances_ != 0) # Fx ê°’ì˜ abs_energy íŠ¹ì§•ë§Œì˜ ìœ ì˜ë¯¸í•¨
```

```python
# â‘¥-9 : Classification reportë¥¼ í†µí•œ ê²€ì¦ ë° ë¶„ì„
from sklearn.metrics import classification_report

classification_report(y_test, xgb_clf.predict(comprehensive_features_test), target_names = ['true', 'false'], output_dict = True)
```
