# 로이터 뉴스 데이터 (2) : 데이터 복원하기

- 원래 텍스트 -> 정수 변환 과정이 필요하나 해당 데이터셋의 경우 이미 되어있음
- 정수 -> 텍스트로 역변환을 통해 데이터를 복원해보자.

```python
word_index = reuters.get_word_index(path="reuters_word_index.json")
```

```python
word_index['the']
# 사실은 4번임
```

```python
word_index['it']
# 사실은 16번임
```

```python
word_index['to']
# 사실은 7번임
```

```python
index_to_word = { index+3 : word for word, index in word_index.items() }
print(index_to_word[4])
print(index_to_word[16])
```

0,1,2번의 경우 자연어 처리에 쓰이는 특별한 토큰들을 위해 맵핑된 번호들이다.
각각 `<pad>`, `<sos>`, `<unk>`라는 토큰을 나타낸다.

이 토큰도 `index_to_word`에 추가해주자.

```python
# index_to_word에 숫자 0은 <pad>, 숫자 1은 <sos>, 숫자 2는 <unk>를 넣어줍니다.
for index, token in enumerate(("<pad>", "<sos>", "<unk>")):
  index_to_word[index]=token

print(' '.join([index_to_word[index] for index in x_train[0]]))
```

## OOV(Out-of-Vocabulary) 문제와 UNK(Unknown)토큰

기계가 알고 있는 단어들의 집합이 Vocabulary인데, 기계가 미처 배우지 못한 단어를 표현한다. 

기계가 모르는 해당 단어들을 전부 `<unk>`라는 토큰으로 변환한다.

아까 데이터 로드할 때, `num_words=10000` -> 10000을 넘는 정수 번호를 가진 단어를 전부 `<unk>`, 우리가 2번으로 할당했기에 2번으로 출력

```python
# 전체 훈련용 뉴스 데이터 + 전체 테스트용 뉴스 데이터를 텍스트 데이터로 변환
decoded = []
for i in range(len(x_train)):
    t = ' '.join([index_to_word[index] for index in x_train[i]])
    decoded.append(t)

x_train = decoded
print(len(x_train))
```

```python
#test데이터도 변환해주세요!

decoded = []
for i in range(len(x_test)):
    t = ' '.join([index_to_word[index] for index in x_test[i]])
    decoded.append(t)

x_test = decoded
print(len(x_test))
```

# 벡터화(Vectorization)



## Bag of Words(BoW)

- 텍스트 데이터의 단어 순서를 무시하고, 출현 빈도만 고려해 문장을 수치화하는 방식.
- 이 가설을 바탕으로 문장들을 행렬로 표현할 것이 바로 **DTM**과 **TF-IDF**

## DTM(Document-Term Matrix)

- 여러 문서(Document)에 등장하는 각 단어의 빈도를 행렬로 나타냄
    - 행 : 각 문서
    - 열 : 전체 문서에 등장하는 고유 단어들
    - 값 : 해당 문서에서 그 단어가 나타난 횟수

## TF-IDF(Term Frequency-Inverse Document Frequency)

- DTM이 'the', 'a' 같은 흔히 등장하지만 큰 의미가 없는 단어에 가중치를 많이 부여한다는 단점을 해결
- TF (단어 빈도) : 특정 문서에서 단어가 많이 나올수록 값이 커진다.
- IDF (역문서 빈도) : 여러 문서에 두루 등장하는 단어일수록 값을 깎는다.

```python
from sklearn.feature_extraction.text import CountVectorizer # DTM
from sklearn.feature_extraction.text import TfidfTransformer # TF-IDF
```

```python
dtmvector = CountVectorizer()
x_train_dtm = dtmvector.fit_transform(x_train)
print(x_train_dtm.shape) # 10000보다 작은 이유 : DTM은 자제적으로 불필요한 토큰을 판단해 제거
```

```python
tfidf_transformer = TfidfTransformer()
tfidfv = tfidf_transformer.fit_transform(x_train_dtm)
print(tfidfv.shape)
```

# 나이브 베이즈 분류기

```python
from sklearn.naive_bayes import MultinomialNB #다항분포 나이브 베이즈 모델
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import ComplementNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score #정확도 계산
```

```python
model = MultinomialNB()
model.fit(tfidfv, y_train)
```

```python
x_test_dtm = dtmvector.transform(x_test) #테스트 데이터를 DTM으로 변환
tfidfv_test = tfidf_transformer.transform(x_test_dtm) #DTM을 TF-IDF 행렬로 변환

predicted = model.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

# F1-Score, Confusion Matrix

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test, model.predict(tfidfv_test), zero_division=0))
```

```python
def graph_confusion_matrix(model, x_test, y_test):#, classes_name):
  df_cm = pd.DataFrame(confusion_matrix(y_test, model.predict(x_test)))#, index=classes_name, columns=classes_name)
  fig = plt.figure(figsize=(12,12))
  heatmap = sns.heatmap(df_cm, annot=True, fmt="d")
  heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=12)
  heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=12)
  plt.ylabel('label')
  plt.xlabel('predicted value')
```

```python
graph_confusion_matrix(model, tfidfv_test, y_test)
```

# 다양한 머신러닝 모델 사용해보기

## 1. Complement Naive Bayes Classifier(CNB)

- 나이브 베이즈가 조건부확률을 사용해 특정 샘플에 가중치가 치우처질 위험이 있다.
- 이러한 클래스 불균형 문제를 고려해 가중치를 부여하는 과정에서 해결하는 것이 CNB

```python
cb = ComplementNB()
cb.fit(tfidfv, y_train)

predicted = cb.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

## 2. 로지스틱 회귀(Logistic Regression)
- SoftMax 함수를 이용한 다중 클래스 분류
> SoftMax : N개의 클래스 분류 문제에서 N차원의 벡터가 각 클래스가 정답일 확률'을 표현하도록 정규화

```python
# 10분 정도 소요됩니다.
lr = LogisticRegression(C=10000, penalty='l2', max_iter=3000)
lr.fit(tfidfv, y_train)

predicted = lr.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

## 3. 선형 서포트 벡터 머신(Linear Support Vector Machine)
- 데이터군 간 분류 시 가장 구분이 잘되는 기준은 어떤 선인가? = 기준선과의 거리(마진)이 가장 큰 기준선


```python
lsvc = LinearSVC(C=1000, penalty='l1', max_iter=3000, dual=False)
lsvc.fit(tfidfv, y_train)
```

```python
predicted = lsvc.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

## 3. 결정 트리(Decision Tree)

```python
tree = DecisionTreeClassifier(max_depth=10, random_state=0)
tree.fit(tfidfv, y_train)

predicted = tree.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

## 4. 랜덤 포레스트(Random Forest)

```python
#estimator를 5로 설정하고 random_state를 0으로 설정한 RandomForestClassifier를 선언해주세요.
#참고자료 : https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html#sklearn.ensemble.RandomForestClassifier
forest = RandomForestClassifier(n_estimators= 5, random_state = 0)
forest.fit(tfidfv, y_train)

predicted = forest.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

## 5. 그래디언트 부스팅 트리(GradientBoostingClassifier)

```python
# 15분 정도 소요될 수 있습니다.
grbt = GradientBoostingClassifier(random_state=0) # verbose=3
grbt.fit(tfidfv, y_train)

predicted = grbt.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```

## 6. 앙상블 - 보팅 (Ensemble Voting)

```python
# 20분 이상 소요될 수 있습니다.
# Voting을 구현해보세요. (Voing은 'soft'로 설정)
# 조건1 : 로지스틱회귀를 사용해주세요(패널티는 'L2'로 설정)
#조건 2: Complement Naive Bayes Classifier를 사용해주세요
#조건 3: GradientBoosting Classifier를 사용해주세요
# 참고자료 : https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingClassifier.html#sklearn.ensemble.VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import ComplementNB
from sklearn.ensemble import GradientBoostingClassifier, VotingClassifier

lr = LogisticRegression(penalty='l2', solver='liblinear', random_state=0)
cb = ComplementNB()
gb = GradientBoostingClassifier(random_state=0)

voting_classifier =  VotingClassifier(voting = 'soft', estimators = [('lr',lr),('cb',cb),('gb',gb)])
voting_classifier.fit(tfidfv, y_train)

predicted = voting_classifier.predict(tfidfv_test) #테스트 데이터에 대한 예측
print("정확도:", accuracy_score(y_test, predicted)) #예측값과 실제값 비교
```
